```python
from decouple import config
import pandas as pd
import sys
import os
sys.path.insert(0, config('PROYECTO_DIR'))
from clases.bd.conexion import  SQLServerConnector
conn = SQLServerConnector('DBGERESA')
pd.set_option('display.max_columns', None)

```

```python
def limpiar_columna(texto): 
    # Reemplazar espacios por guiones bajos
    texto_sin_espacios = texto.replace(' ', '_')
    # Quitar los puntos
    return texto_sin_espacios.replace('.', '')


# Ruta de la carpeta que contiene los archivos .xlsx
folder_path = os.path.join(config('PROYECTO_DATA'), "REFCON/")

# Lista para almacenar los DataFrames cargados
dataframes = []

# Iterar sobre todos los archivos .xlsx en la carpeta
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):  # Solo archivos .xlsx        
        file_path = os.path.join(folder_path, filename)
        
        try:
            # Intentar cargar el archivo .xlsx a un DataFrame de Pandas
            df = pd.read_excel(file_path, engine='openpyxl')  # Usamos el motor 'openpyxl' para leer archivos .xlsx
            df.columns = [limpiar_columna(col) for col in df.columns]
            if filename.startswith("trama_recibida"):
                print(filename)                
                df.rename(columns={"MRED_ORIGEN": '2',"COD_UNICO_ORIGEN": '3'}, inplace=True)
                df.rename(columns={"2": 'COD_UNICO_ORIGEN',"3": 'MRED_ORIGEN'}, inplace=True)

            # Seleccionar solo las columnas deseadas
            columns_to_select = ['N°', 'NRO_REFERENCIA', 'NRO_HC', 'AÑO', 'MES', 'FECHA_REGISTRO',
       'FECHA_ENVIO', 'FECHA_ACEPTACION', 'FECHA_RECIBIDO', 'FECHA_OBSERVADO',
       'FECHA_RECHAZADO', 'FECHA_CITA', 'FECHA_CONTRAREFERIDO', 'DISA_ORIGEN',
       'RED_ORIGEN', 'MRED_ORIGEN', 'COD_UNICO_ORIGEN','IPRESS_ORIGEN',
       'DISA_DESTINO', 'RED_DESTINO', 'MRED_DESTINO', 'COD_UNICO_DESTINO',
       'IPRESS_DESTINO', 'TIPO_DOC', 'NRO_DOC', 'NOMBRE_PACIENTE',
       'EST_PACIENTE', 'FECHA_NACIMIENTO', 'EDAD', 'TIPO_EDAD', 'SEXO',
       'ID_CORREL', 'COD_CIEX/CPT', 'DES_CIEX/CPT', 'TIPO_DIAG',
       'EST_REFERENCIA', 'COMENTARIO_ESTADO', 'MOTIVO', 'FINANCIADOR',
       'UPS_ORIGEN', 'UPS_DESTINO', 'ESPECIALIDAD', 'TIPO_DOC_RESPONSBLE_REF',
       'NUMERO_DOC_RESPONSBLE_REF', 'NOMBRE_PERSONAL_RESPONSBLE_REF',
       'TIPO_DOC_REGISTRADOR', 'DOCUMENTO_REGISTRADOR', 'PERSONAL_REGISTRADOR',
       'TIPO', 'LLEGO_AL_EESS', 'PERSONAL_ACEPTA', 'PERSONAL_CITA',
       'PERSONAL_RECIBE', 'PERSONAL_OBSERVA', 'PERSONAL_RECHAZA',
       'PERSONAL_ANULA']
            df = df[columns_to_select]  # Filtrar las columnas que necesitas

            # Añadir el DataFrame a la lista
            dataframes.append(df)
        
        except Exception as e:
            print(f"Error cargando el archivo {filename}: {e}")

# Si quieres combinar todos los DataFrames en uno solo (opcional)
combined_df = pd.concat(dataframes, ignore_index=True)

# Mostrar las primeras filas del DataFrame combinado
combined_df.head()
```
