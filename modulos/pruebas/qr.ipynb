```python
import qrcode

# URL que quieres convertir en QR
url = "https://drive.google.com/drive/folders/1SZaNfsR5aBk03iyYI7wFMpzRyRPytRt7"

# Crear el código QR
qr = qrcode.QRCode(
    version=1,            # Tamaño del QR (1 a 40)
    error_correction=qrcode.constants.ERROR_CORRECT_L,  # Nivel de corrección de errores
    box_size=20,          # Tamaño de cada cuadro
    border=2,             # Grosor del borde
)

qr.add_data(url)
qr.make(fit=True)

# Crear una imagen del QR
img = qr.make_image(fill_color="black", back_color="white")

# Guardar la imagen en un archivo
img.save("codigo_qr.png")

#print("Código QR generado y guardado como 'codigo_qr.png'")
from IPython.display import display
display(img)
```
